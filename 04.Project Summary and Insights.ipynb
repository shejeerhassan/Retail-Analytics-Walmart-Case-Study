#### 4.1 Project Summary
  
Analyzed sales, profit, discounts, margins, shipping cost, product categories, regions, and customer segments to identify revenue and profitability drivers and recommend actions to improve margins and operational efficiency


#### 4.2🔑 Five Key Insights
- Sales are uneven → Most transactions are small, with a few very large orders pulling up the average.
- Margins vs Profit contribution mismatch → High‑margin sub‑categories do not always deliver the highest profit (Appliances, Binders, Chairs); profit leaders are Telephones, Office Machines, and Binders.
- Shipping costs vary sharply by category → Furniture is most expensive to ship, while Office Supplies are cheapest. Larger sales orders tend to incur higher shipping costs.
- Regional and segment performance → Corporate customers in the Central region dominate sales; Technology category yields the highest average profit.
- Discounts have little impact on sales → Sales remain highest at low discounts (0–10%), and discounts slightly erode profit without driving volume.


#### 4.3 Key Metrics
- Mean Sales per transaction: 1775.88
- Average Profit per sale: 181.18
- Highest Sales Sub‑Category: Office Machines
- Top Customer Segment: Corporate group
- Correlation (Sales vs Profit): 0.58 (moderate positive)
- Correlation (Sales vs Shipping Cost): 0.43 (moderate positive)
- Correlation (Sales vs Discount): -0.02 (negligible)


#### 4.4 Recommendations
**1. Focus on profitable categories**   
Focus sales, marketing, and inventory planning on high‑profit categories such as Telephones, Office Machines, and Technology. These consistently deliver strong profit contributions. Streamline resources away from low‑profit or negative‑contribution categories (e.g., Tables, Bookcases) to maximize ROI.   

**2.Optimize logistics for high‑cost categories**   
Furniture incur the highest shipping costs. Negotiate better carrier contracts, explore bulk or consolidated shipping, and consider dynamic pricing that factors in shipping expenses. This will reduce logistics overhead and protect margins in categories with heavy shipping burdens.

**3.Refine discount strategy**   
Avoid discounts which not boosting sales. Instead, use targeted promotions for attracting new customer segments, seasonal campaigns, or slow‑moving products. This ensures discounts are strategic tools for clearing inventory or attracting new customers, rather than reducing overall profitability

**4.Target corporate customers in Central region**   
Corporate customers in the Central region are the strongest revenue drivers. Expand tailored offers, loyalty programs, and dedicated customer management for this segment. Strengthening relationships here can secure repeat business and long‑term contracts, stabilizing revenue streams.

**5. Align margin and profit contribution**   
Some high‑margin categories underperform in realized profit. Conduct a detailed study of costs, revenue , discounts and shippinh cost to identify leakage points. Make relevant pricing strategies, promotional policies, or product bundling to ensure that high margins translate into actual profit contribution.


**Takeaway**   
By focusing on profitable categories, tightening discount policies, optimizing shipping, and leveraging strong customer segments, the business can improve efficiency, protect margins, and drive sustainable growth.


**Conclusion**   

The analysis reveals that profitability is driven more by product mix and customer segment than by discounts. While sales and profit are moderately linked, discounts are not a strong driver of performance. Shipping costs remain a critical factor, especially for Furniture. By focusing on profitable categories, optimizing logistics, and refining discount strategies, the business can improve overall efficiency and profitability.
