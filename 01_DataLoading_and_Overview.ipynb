**1.1 Dataset Overview**   
___
This dataset captures five years of Walmart retail transactions, providing details on customers, orders, products, and logistics. Key information includes:
- Customer data: demographics (age, name, segment), location (city, state, ZIP).
- Order details: order ID, date, priority, quantity, discounts, and shipping (mode, cost, dates).
- Product attributes: category, sub-category, name, container type, unit price, and base margin.
- Performance metrics: sales, profit, and regional breakdowns.
  
Together, these fields enable analysis of customer behavior, product performance, profitability, and operational efficiency across Walmart’s retail ecosystem.

**1.6 Data Quality Checks**   
To check how many missing values and duplicate items exist in data

```python
# Check for missing values in all columns
print("Missing Values Summary:")
missing_values = df.isnull().sum()
print(missing_values)
print("="*70)

# Columns with missing values
if missing_values.sum() > 0:
    print("\nColumns with missing values:")
    print(missing_values[missing_values > 0])
else:
    print("\nNo missing values found in the dataset.")
```

```python
# Check for duplicate rows
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")

if duplicates > 0:
    print(f"\nWarning: {duplicates} duplicate row(s) detected!")
else:
    print("\nNo duplicate rows found in the dataset.")
```
